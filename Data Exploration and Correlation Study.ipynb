# Data Exploration and Correlation Study

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds = pd.read_csv('employees_data.csv')
```

### 3 - Display the average of these attributes: ‘Age’, ‘Monthly Income’ and ‘Years at Company’

```python
print('Average value of attribute "Age" =', round(np.average(ds['Age']), 2))
print('Average value of attribute "Monthly Income" =', round(np.average(ds['MonthlyIncome']), 2))
print('Average value of attribute "Years at Company" =', round(np.average(ds['YearsAtCompany']), 2))
```

### 4 - Find the minimum and maximum ‘Monthly Income’

```python
print('Minimum value of attribute "Monthly Income" =', min(ds['MonthlyIncome']))
print('Maximum value of attribute "Monthly Income" =', max(ds['MonthlyIncome']))
```

### 5 - Provide graphical visualization by plotting Histogram of ‘Monthly Income vs staff numbers’

```python
plt.hist(ds['MonthlyIncome'])
plt.title('Histogram of Monthly Income vs Staff Number')
plt.xlabel('Monthly Income')
plt.ylabel('Staff Number')
plt.show()
```

### 6 - Provide graphical Visualization of the distribution between ‘Year at Company’ and ‘Monthly Income’ using the scatter plot

```python
plt.scatter(ds['YearsAtCompany'], ds['MonthlyIncome'])
plt.title('Monthly Income vs Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Monthly Income')
plt.show()
```

### 7 - Find the correlation between ‘Years at Company’ and ‘Monthly Income’

```python
corr = ds[['YearsAtCompany', 'MonthlyIncome']].corr()
print(corr)
```

```python
import seaborn as sns

sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
plt.show()
```

### 8 - Brief discussion on: a) Range of monthly income at Company A, b) Most and Least Frequent monthly income values at Company A, c) Distribution of monthly income values, d) The relationship between monthly income and employees’ individual years at the company

```python
# Calculate the range for "MonthlyIncome"

max(ds['MonthlyIncome'])-min(ds['MonthlyIncome'])
```

```python
# Get the mode for "MonthlyIncome"

from scipy import stats
stats.mode(ds['MonthlyIncome'])
```

```python
# Get the count of each unique value under column "MonthlyIncome" in ascending order
# Which means the list will start with the value(s) with the lowest count (i.e., least frequent)  

pd.set_option('display.max_rows', None)
ds['MonthlyIncome'].value_counts(ascending=True)
```

```python
# Find the median for "MonthlyIncome" 

np.median(ds['MonthlyIncome'])
```

```python
# Find the mean for "MonthlyIncome"
# The value should be the same as the average value calculated previously

np.mean(ds['MonthlyIncome'])
```

### Discussion summary:

1 - Range of monthly income is 18990.

2 - Most frequent monthly income value is 2342, with the count of 4. There are many least frequent values for monthly income, all sharing the same number of occurence of 1, as shown listed above. The last value recorded (refer to the top of the ascending list above) is 14336.

3 - Looking at the histogram plotted in Section A5, rough observation is that the number of staffs gets higher as the monthly income decreases. It is observed that the distribution is right-skewed (i.e., positively skewed), which suggests that mean > median > mode. This deduction is further supported by the values calculated previously: mean (6502.93) > median (4919) > mode (2342).

4 - Looking at the scatter plot in Section A6, linear relationship is not quite visible between monthly income and years at company. This is futher supported by the correlation (corr) value calculated in Section A7, where corr = 0.51. The value is right at the middle in between 0 and 1, suggesting that the correlation is low. It is neither perfect/strong nor no correlation, hence, the not-quite-visible linear relationship. The correlation value is positive, indicating that monthly income and years at company basically fluctuate in the same direction (i.e., monthly income increases as years at company increases, and vice versa). From a logical point-of-view, this is quite sensible given that longer experience/service years typically translates to higher salary.
